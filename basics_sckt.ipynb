```python
#libraries import
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#simple linear regression
from sklearn.linear_model import LinearRegression
```

```python
#load the data
df=sns.load_dataset("tips")
df.head()
```

```python
sns.scatterplot(x='total_bill',y='tip',data=df)
```

```python
#create a regression line plot using seaborn
sns.lmplot(x='total_bill',y='tip',data=df)
```

```python
#select the feature column from df
x= df[['total_bill']]#double dimension
#select the target7 column from df
y=df['tip']
```

```python
#load linear regression
model=LinearRegression()
#fit the model
model.fit(x,y)#means train the model
```

```python
#lets predict the tip foor an unknown total_bill
#total bill=50
model.predict([[70]])
```

```python
#write the whole script in one go
#libraries import
import pandas as pd
from sklearn.linear_model import LinearRegression
#load the data
df=sns.load_dataset('tips')
#select the feature column from df
x=df[['total_bill']]
#select the target column from df
y=df['tip']
#load linear regression
model=LinearRegression()
#fit the model
model.fit(x,y)#means train the model
#lets predict the tip foor an unknown total_bill
#total bill=50
model.predict([[70]])
```
